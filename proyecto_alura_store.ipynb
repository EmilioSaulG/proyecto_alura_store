```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.head()
```

```python
tienda3.head()
```

```python
tienda4.head()
```

```python
tienda['tienda'] = 'Tienda 1'
tienda2['tienda'] = 'Tienda 2'
tienda3['tienda'] = 'Tienda 3'
tienda4['tienda'] = 'Tienda 4'
```

```python
df = pd.concat(
    [tienda, tienda2, tienda3, tienda4],
    ignore_index=True
)
```

```python
df.head()
df.info()
df['tienda'].value_counts()
```

```python
df.shape
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

## **1. Análisis de facturación**

```python
#Con este generamos el promedio de cada tienda
ingreso_por_tienda = df.groupby('tienda')['Precio'].sum()
ingreso_por_tienda
```

```python
#Generamos un grafico de ventas por tienda
ingresos_por_tienda = df.groupby('tienda')['Precio'].sum()

plt.figure()
plt.bar(ingresos_por_tienda.index, ingresos_por_tienda.values)
plt.title('Ingresos totales por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.show()
```

## **2. Ventas por categoría**

```python
#Ventas por categoría en cada tienda
df.groupby(['tienda','Categoría del Producto']).size().reset_index(name='Ventas')
```

```python
#Generamos gráfico de ventas por categoría.
ventas_categoria = df['Categoría del Producto'].value_counts().head(10)

plt.figure()
plt.barh(ventas_categoria.index, ventas_categoria.values)
plt.title('Las 10 categorías más vendidas')
plt.xlabel('Número de ventas')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
plt.show()
```

## **3. Calificación promedio de la tienda**

```python
calificacion_tienda = df.groupby('tienda')['Calificación'].mean()
calificacion_tienda
```

```python
#Generamos el promedio de calificacion de cada tienda
calificacion_promedio = df.groupby('tienda')['Calificación'].mean()

plt.figure()
plt.bar(calificacion_promedio.index, calificacion_promedio.values)
plt.title('Calificación promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Promedio')
plt.show()
```

## **4. Productos más y menos vendidos**

```python
#Porductos mas vendidos
top_productos = df['Producto'].value_counts().head(10)
top_productos
```

```python
#Generamos gráfico de los productos mas vendidos
plt.figure()
plt.barh(top_productos.index, top_productos.values)
plt.title('Top 10 productos más vendidos')
plt.xlabel('Número de ventas')
plt.ylabel('Producto')
plt.gca().invert_yaxis()
plt.show()
```

```python
#Productos menos vendidos
bottom_productos = df['Producto'].value_counts().tail(10)
bottom_productos
```

```python
#Generamos gráfico de los porductos menos vendidos
plt.figure()
plt.barh(bottom_productos.index, bottom_productos.values)
plt.title('Top 10 productos menos vendidos')
plt.xlabel('Número de ventas')
plt.ylabel('Producto')
plt.gca().invert_yaxis()
plt.show()
```

## **5. Envío promedio por tienda**

```python
#Generamos el promedio de envio de cada tienda
envio_promedio = df.groupby('tienda')['Costo de envío'].mean()
envio_promedio
```

```python
#Resumen estadístico de precios y envíos
df[['Precio','Costo de envío']].describe()
```

```python
#Generamos el gráfico de envío promedio por tienda
plt.figure()
plt.bar(envio_promedio.index, envio_promedio.values)
plt.title('Envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo por envío Promedio')
plt.show()
```
